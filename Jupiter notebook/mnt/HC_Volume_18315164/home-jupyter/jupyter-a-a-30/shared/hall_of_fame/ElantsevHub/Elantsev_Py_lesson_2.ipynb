```python
# Загружаем библиотеки
import pandas as pd
```

```python
# Загружаем базу данных
bookings = pd.read_csv('../../homeworks/python_ds_miniprojects/2/bookings.csv', sep=';')
bookings.head(7)
```

```python
# Приводим названия колонок в удобный для работы формат
bookings.columns = bookings.columns.str.replace(" ","_").str.lower()
bookings.head(7)
```

```python
# Какие типы данных есть в таблице
bookings.dtypes.value_counts()
```

```python
# Список названаий всех колонок, для удобства в работе
bookings.columns
```

```python
# Топ-5 стран из которых совершено бронирование
bookings.query("is_canceled == 0") \
        .country \
        .value_counts() \
        .head(5)
```

```python
# Средняя длительность бронирования отелей разных типов
bookings.groupby('hotel') \
        .agg({'stays_total_nights': 'mean'}) \
        .round(2)
```

```python
# За какие годы есть данные в базе
bookings.arrival_date_year.value_counts()
```

```python
# На какой месяц чаще всего оформляли бронь в 2016 году?
bookings.query("arrival_date_year == 2016") \
        .arrival_date_month \
        .value_counts()
```

```python
# На какой месяц чаще всего оформляли бронь в 2017 году?
bookings.query("arrival_date_year == 2017") \
        .arrival_date_month \
        .value_counts()
```

```python
# На какой месяц (arrival_date_month) бронирования отеля типа City Hotel отменялись чаще всего в 2015? 2016? 2017? 
bookings.query("hotel == 'City Hotel' and is_canceled == 1") \
        .groupby('arrival_date_year') \
        .arrival_date_month \
        .value_counts()
```

```python
# Какая из колонок [adults, children, babies] имеет наибольшее среднее значение?
bookings[['adults', 'children', 'babies']].describe()
```

```python
# В отелях какого типа в среднем больше детей?
# Сперва создаем колонку total_kids - в которой вычисляем общее количество детей
bookings['total_kids'] = bookings.children + bookings.babies

# Вычисляем среднее количество детей по типу отеля
bookings.groupby('hotel', as_index=False) \
        .agg({'total_kids': 'mean'}) \
        .round(2) \
        .sort_values('total_kids', ascending=False)
```

```python
# Как Churn_rate (отток, коэффициент оттока) связан с наличием детей у клиентов
# Сперва создадим колонку has_kid
bookings['has_kid'] = bookings.total_kids > 0

bookings['has_kid']
```

```python
# Churn_rate для посетителей без детей
bookings.query("has_kid == False") \
        .is_canceled \
        .value_counts(normalize=True) \
        .mul(100) \
        .round(2)
```

```python
# Churn_rate для посетителей с детьми
bookings.query("has_kid == True") \
        .is_canceled \
        .value_counts(normalize=True) \
        .mul(100) \
        .round(2)
```
