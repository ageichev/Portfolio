К вам пришёл коллега с [немного обновленными данными](https://stepik.org/media/attachments/lesson/361624/transaction_data_updated.csv). Постройте сводную таблицу `user_vs_minute_pivot`, где:

- столбцы – имена пользователей
- строки – минута времени осуществления операции (minute)
- значения – число совершённых операций

*Пропущенные значения заполните нулями.*

*Note:* фильтровать успешные транзакции не нужно - постройте табличку по всем

Новые данные сохранены в `df2`. 

```python
df2 = pd.read_csv('transaction_data_updated.csv', parse_dates=['date'])
df2.head()
```

```python
user_vs_minute_pivot = (df2
                         .groupby(['name', 'minute'], as_index=False)
                         .size()
                         .pivot(index='minute', columns='name', values='size')
                         .fillna(0)
                       )
user_vs_minute_pivot.shape
```

## 7. Исправьте ошибку коллеги (если она есть 😉)
- Посчитайте число минут, прошедших с начала дня. Результат сохраните в колонку `true_minute`.
- Постройте график, как в предыдущем шаге, и посмотрите, произошли ли изменения (график в код включать не нужно).

**Например:**


Данные записаны в `df2` , колонка `date` прочитана как дата. 

```python
df2['true_minute'] = df2.date.dt.hour.mul(60) + df2.date.dt.minute
df2.head()
```

```python
data = (df2
        .true_minute
        .value_counts()
        .reset_index()
        .rename(columns={'index': 'true_minute',
                         'true_minute': 'count'})
        .sort_values('true_minute')
        .reset_index(drop=True)
       )
data.dtypes
```

```python
d = (df2
    .true_minute
    .value_counts()
    .sort_index()
)
```

```python
sns.barplot(x=d.index, y=d.values)
```
