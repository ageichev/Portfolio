```python
df_time=[]
for i in ways:
    file=pd.read_csv(i)
    f=i.split('/')
    file['name']=f[-2]
    file['date']=f[-3]
    df_time.append(file)
```

```python
df_time
```

```python
big_df=pd.concat(df_time)
```

```python
big_df
```

```python
big_df.quantity.sum()
```

```python
big_df.drop(columns=['Unnamed: 0']).reset_index(drop=True)
```

```python
max_by=big_df.groupby('name',as_index=False).agg({'quantity' :'sum'}).sort_values('quantity', ascending=False)
```

```python
max_by.head()
```

```python
top_product=big_df.groupby('product_id', as_index=False).agg({'quantity':'sum'}).sort_values('quantity', ascending=False)
```
