```python
import pandas as pd

df = pd.DataFrame({'Дата': ['01.01.2022', '01.02.2022', '01.03.2022', '01.04.2022'],
                   'Начало поездки': ['Москва', 'Питер', 'Новосибирск', 'Екатеринбург'],
                   'Конец поездки': ['Питер', 'Новосибирск', 'Екатеринбург', 'Казань'],
                   'Расстояние (км)': [800, 1500, 2000, 700],
                   'Стоимость (руб)': [10000, 20000, 25000, 7000]})
print(df)
```

```python
import matplotlib.pyplot as plt

# создаем график
plt.plot(df['Дата'], df['Стоимость (руб)'])

# добавляем название осей
plt.xlabel('Дата')
plt.ylabel('Стоимость (руб)')

# добавляем название графика
plt.title('Стоимость поездок такси')

# добавляем легенду
plt.legend(['Стоимость'])

# показываем график
plt.show
```

```python
from itertools import combinations

def binomial_coefficient(n, k):
    return len(list(combinations(range(n), k)))

print(binomial_coefficient(50, 5))
```
